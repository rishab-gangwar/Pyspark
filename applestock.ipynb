```
spark=SparkSession.builder.appName('date').getOrCreate()
```

```
df=spark.read.csv('dbfs:/FileStore/shared_uploads/195103@nith.ac.in/appl_stock.csv',header=True,inferSchema=True)
```

```
df.show()
```

```
df.head(1)
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import dayofmonth,hour,dayofyear,month,year,weekofyear,format_number,date_format
```

```
df.select(year(df['date'])).show()
```

```
new=df.withColumn("Year",year(df['Date']))
```

```
res=new.groupBy("Year").mean().select(["Year","avg(Close)"])
```

```
new=res.withColumnRenamed("avg(Close)","avg closing price")
```

```
new.select(['Year',format_number('avg closing price',2)]).show()
```
